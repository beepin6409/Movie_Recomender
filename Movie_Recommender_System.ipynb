```python
#since it is a string of list , indexing cant be done like list so it must be converted to list first 
#in python there is a module used to convert string of list to list ast and function name is literal_eval()
```

```python
import ast
```

```python
ast.literal_eval(movies['genres'].iloc[0])      # now we get the  list which can be indexed.
```

```python
 def convert(obj):
        L=[]
        for i in ast.literal_eval(obj):
            L.append(i['name'])
        return L
```

```python
convert(movies['genres'].iloc[1])
```

```python
#converting the every genres into list 
movies['genres']=movies['genres'].apply(convert)
```

```python
#Since in the cast there are a lot of actors so taking only 4 names of them.
def convert_cast(obj):
        L=[]
        count=0
        for i in ast.literal_eval(obj):
            L.append(i['name'])
            count=count+1;
            if(count==4):
                break
        return L
```

```python
movies['cast']=movies['cast'].apply(convert_cast)
```

```python
movies.head(10)
```

```python
movies['crew'].iloc[0]
```

```python
def convert_crew(obj):
    direct=[]
    for i in ast.literal_eval(obj):
        if(i['job']=='Director'):
            direct.append(i['name'])
            break
    return direct
```

```python
convert_crew(movies['crew'].iloc[0])
```

```python
movies['crew']=movies['crew'].apply(convert_crew)
```

## Stemming is done to remove similar or same words with plural form or ing , etc  
## For eg action, actions are same so making both of them action only 

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
stem(final_df['tags'][0])
```

```python
final_df['tags']=final_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(final_df['tags']).toarray()
```

```python
vectors
```

```python
vectors.shape
```

```python
cv.get_feature_names()
```

```python
#calculating the cosine distance between each movies
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
similarity.shape
```

```python
#enumerate function is used to add index  , with values and give of tuples containing (index,value)
list(enumerate(similarity[0]))
```

```python
sorted(list(enumerate(similarity[0])),key=lambda x:x[1],reverse=True)  #sorting in descending order on the basis of similarity value 
```

```python
def recommend(movie):
    movie_index=final_df[final_df['title']==movie].index[0]
    distance=similarity[movie_index]
    movies_list = sorted(list(enumerate(distance)),key=lambda x:x[1],reverse=True)[1:7]
    
    for i in movies_list:
        print(final_df['title'][i[0]])
```

```python
recommend('Independence Day')
```
